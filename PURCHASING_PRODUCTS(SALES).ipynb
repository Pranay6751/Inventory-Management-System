# READING ITEMS FROM THE SALES JSON

```python
import json
f1=open("sales.json","r")
txt=f1.read()
f1.close()
```

```python
txt
```

```python
sale=json.loads(txt)
sale
```

```python
import time
while(True):
  print("********************")
  print("Welcome to the Shopping Mall")
  print("********************\n")
  total_amt=0
  prod_id_list=[]
  prod_name_list=[]
  price_list=[]
  exp_date_list=[]
  time_list=[]
  quantity_list=[]
  bill_prod_list=[]
  my_cart={}
  j=1
  i=len(sale)+1
  while(True):

    print("1.Buy an item\n2.Remove an item from my cart\n3.My cart details\n4.My total Billing Amount")
    opt=int(input("Enter your choice : "))
    if(opt==1):
      x=input("Enter the product id to be purchased : ")
      if(x in record.keys()):
        q=int(input("Enter the number of items to be purchased : "))
        if(q<=record[x]['quantity'] and q>=0):
          print("You succesfully purchased ",q," items of ",record[x]['product name'])
          record[x]['quantity']-=q
          #sale[i]={'product name':,'price':record[x]['price'],'quantity':q,'time':time.ctime(),'billing amount':record[x]['price']*q} 
          prod_id_list.append(x)
          prod_name_list.append(record[x]['product name'])
          price_list.append(record[x]['price'])
          exp_date_list.append(record[x]['expiry date'])
          time_list.append(time.ctime())
          quantity_list.append(q)
          bill_prod_list.append(record[x]['price']*q)

          #Adding items to the cart

          my_cart[x]={'product name':record[x]['product name'],'price':record[x]['price'],'time':time.ctime(),'quantity':q,'expiry date':record[x]['expiry date'],'billing amount':record[x]['price']*q}
          print('\n********************')
          print('Product id : ',x)
          print('Product purchased : ',record[x]['product name'])
          print("Price : ",record[x]['price'])
          print('Expiry date : ',record[x]['expiry date'])  
          print('Time : ',time.ctime())
          print('Quantity : ',q)
          print('********************')
          print('Amount billing on this product: ',record[x]['price']*q,"₹")
          total_amt+=record[x]['price']*q;  
          print('********************\n') 
        else:
          print("Not sufficient ,we have only ",record[x]['quantity']," number of quantities")
      else:
        print("We have no items with that product id, So please enter a correct product id ")

    elif(opt==2):
      x=input("Enter the product id to be removed from the cart : ")
      if(x in record.keys()):
        ind=prod_id_list.index(x)
        prod_id_list.pop(ind)
        prod_name_list.pop(ind)
        exp_date_list.pop(ind)
        time_list.pop(ind)
        quantity_list.pop(ind)
        bill_prod_list.pop(ind)
        del my_cart[x]
        total_amt=total_amt-price_list[ind]
        price_list.pop(ind)
        print("You successfully removed the item from cart\n")
      else:
        print("We have no items with that product id, So please enter a correct product id ")

    elif(opt==3):
      print("Cart Details are")
      print(my_cart)

    elif(opt==4):
      print("\n********************")
      print("Total Amount you spent : ",total_amt,"₹")
      print("********************\n")
    print("Do you want to continue, Enter yes to continue else no")
    if(input()=="yes"):
      continue
    else:
      print("\n********************")
      print("Thank you visit again")
      print("********************")
      print("***")
      print("***")
      break
  sale[i]={'product id':prod_id_list,'product name':prod_name_list,'price':price_list,'quantity':quantity_list,'exp_date':exp_date_list,'time':time_list,'bill on each product':bill_prod_list,'total amount spent':total_amt}
  print("Is there any more customers, Enter yes to continue else no")
  if(input()=="yes"):
    continue
  else:
    break
  


#updating the sale json
js=json.dumps(sale)
f1=open("sales.json",'w')
f1.write(js)
f1.close()
```
