# READING DATA FROM RECORDS JSON

```python
import json

fd=open('records.json','r')
txt=fd.read()
fd.close()

record=json.loads(txt)

```

```python
record
```

#ADDING NEW ITEMS INTO INVENTORY

```python
n=int(input("Enter the number of new items to be added to the inventory : "))
i=1
print()
while(i<=n):
  print("**********************")
  print("Enter item ",i," details")
  prod_id=input("Enter product id : ")
  prod_name=input("Enter product name : ")
  price=int(input("Enter the price(₹) : "))
  quant=int(input("Enter the quantity : "))
  exp_date=input("Enter the expiry date[DD/MM/YYYY] : ")
  mfd_date=input("Enter the manufacture date[DD/MM/YYYY] : ")
  record[prod_id]={'product name':prod_name,'price':price,'quantity':quant,'expiry date':exp_date,'mfd':mfd_date}
  i+=1
  print("**********************")
  print()
print("**********************")
print("you successfully entered ",n," number of items into the inventory")
print("**********************")
js=json.dumps(record)
fd=open("records.json","w")
fd.write(js)
fd.close()
```

# UPDATING THE EXISTING ITEMS

```python
i=1
while(True):
  print("\n**********************")
  prod_id=input("Enter the product id in which quantity is to be updated : ")
  print("product name : ",record[prod_id]['product name'])
  print("price : ",record[prod_id]['price'])
  print("Before updating the quantity is : ",record[prod_id]['quantity'])
  quant=int(input("Enter the number of new items to be added : "))
  record[prod_id]['quantity']=record[prod_id]['quantity']+quant
  print("After updating, the quantity is : ",record[prod_id]['quantity'])
  print("\n**********************\n")
  print("Successfully Updated\n")
  i+=1
  print("Do you want to update some more items\n Enter yes to continue else no")
  if(input()=="yes"):
    continue
  else:
    break
print("\nSuccessfully updated ",i," items")
#Updating the record json file

js=json.dumps(record)
fd=open("records.json","w")
fd.write(js)
fd.close()
```
